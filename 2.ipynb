```python
import pandas as pd
from sklearn.datasets import load_iris
#Membaca file survey lung cancer.csv
iris = pd.read_csv('survey lung cancer.csv')
```

```python
# Melihat informasi dataset
iris.info()

# melihat informasi dataset pada 5 baris pertama
iris.head()
```

```python
iris.columns = ["Gender", "Age", "Smoking", "Yellow Fingers", "Anxiety", "Peer Pressure",
              "Chronic Disease", "Fatigue", "Allergy", "Wheezing", "Alcohol Consuming", "Coughing", "Shortness of Breath", "Swallowing Difficulty", "Chest Pain", "Lung Cancer"]
```

```python
#transforming dataset with dummies variables to replace characters with binary integers
iris["Gender"].replace(["M","F"], [1,2], inplace=True)

#memisahkan atribut dan label
X = iris[["Gender", "Age", "Smoking", "Yellow Fingers", "Anxiety", "Peer Pressure", "Chronic Disease", "Fatigue", "Allergy", "Wheezing", "Alcohol Consuming", "Coughing", "Shortness of Breath", "Swallowing Difficulty", "Chest Pain"]]
y = iris["Lung Cancer"]

#Membagi dataset menjadi data latih & data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
iris.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

#membuat model Decision Tree
tree_model = DecisionTreeClassifier()

#melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)

#Evaluasi Model
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

y_pred = tree_model.predict(X_test)

acc_secore = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_secore)
print('classification report')
print(classification_report(y_pred, y_test))
```

```python
# prediksi model degnan tree_model.predict([["Gender", "Age", "Smoking", "Yellow Fingers", "Anxiety", "Peer Pressure", "Chronic Disease", "Fatigue", "Allergy", "Wheezing", "Alcohol Consuming", "Coughing", "Shortness of Breath", "Swallowing Difficulty", "Chest Pain"]])
print(tree_model.predict([[1, 63, 1, 2, 1, 2, 1, 2, 2, 1, 1, 2, 2, 1, 2]])[0])
```
